```python
import twint
import nest_asyncio
import pandas as pd
import time
import json
nest_asyncio.apply()
```

```python
pd.set_option('display.max_colwidth', None)
pd.set_option('display.max_columns', None)
PROJECT_NAME = "Kim Seon Ho 2"
DATASET_PATH = "/content/drive/MyDrive/Data/Dataset/KimSeonHo"
```

```python
!cp -r "$DATASET_PATH" .
```

```python
# Configure
c = twint.Config()
c.Store_json = True
c.Output = f"{PROJECT_NAME}_new.json"
c.Search = '"#KimSeonHo" OR "Kim Seonho" OR "kim seon" OR "Kim Seon-ho" OR "Seonho" OR "dusik"'
c.Lang = "id"
c.Since = "2021-10-19"
# c.Until = "2021-10-25"
c.Resume = f'{PROJECT_NAME}_resume.txt' 
c.Hide_output = True
```

```python
while True:
  try:
    twint.run.Search(c)
  except:
    time.sleep(1)
    print('Error. Restarting...')
  else:
    print("Finished")
    break
```

```python
data_new = []
for line in open(f"{PROJECT_NAME}_new.json", 'r'):
    data_new.append(json.loads(line))
```

```python
df_new = pd.DataFrame(data_new)
df_new["created_at"] = pd.to_datetime(df_new["created_at"]).dt.tz_convert('Asia/Jakarta')
df_new["date"] = pd.to_datetime(df_new["date"])
df_new["id"] = pd.to_numeric(df_new["id"])
```
